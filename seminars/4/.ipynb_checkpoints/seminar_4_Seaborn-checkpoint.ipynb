Мы подробно рассмотрим пакет Matplotlib для визуализации в Python.
Matplotlib - это многоплатформенная библиотека визуализации данных, построенная на массивах NumPy и предназначенная для работы с более широким стеком SciPy.

Одной из важнейших особенностей Matplotlib является его способность хорошо работать со многими операционными системами и графическими платформами. Matplotlib поддерживает десятки бэкендов и типов вывода, что означает, что вы можете рассчитывать на его работу независимо от того, какую операционную систему вы используете или какой формат вывода вы хотите. 
Этот кросс-платформенный подход "все для всех" был одной из самых сильных сторон Matplotlib. 
Он привел к большой базе пользователей, что в свою очередь привело к активной базе разработчиков, мощным инструментам Matplotlib и повсеместному распространению в научном мире Python.

## Два интерфейса по цене одного

Потенциально запутанной особенностью Matplotlib является наличие двойного интерфейса взаимодействия: 

- легкий интерфейс в стиле MATLAB, основанный на состояниях
- мощный объектно-ориентированный интерфейс

Рассмотрим различия между ними.

#### MATLAB-style Interface

````python
plt.figure()  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
````

#### Object-oriented interface

````python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
````

## Simple Line Plots (Простые графики)

````python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np

fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
````

````python
plt.plot(x, np.sin(x - 0), color='blue'       ) # specify color by name
plt.plot(x, np.sin(x - 1), color='g'          ) # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75'       ) # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44'    ) # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse' ) # all HTML color names supported
````

````python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
````

````python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
````

## Scatter Plots (График рассеяния)

````python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
````

````python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black');
````

````python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
````

````python
plt.plot(x, y, '-ok');
````

````python
plt.plot(x, y, '-p', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2);
````

````python
plt.scatter(x, y, marker='o');
````

````python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar();  # show color scale
````

````python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
````

### Выводы:

Помимо различных функций, доступных в ``plt.plot`` и ``plt.scatter``, почему вы можете предпочесть один из них другому? 
Для небольших объемов данных это не имеет большого значения, когда наборы данных становятся больше нескольких тысяч точек, ``plt.plot`` может быть заметно более эффективным, чем ``plt.scatter``.
Причина в том, что ``plt.scatter`` имеет возможность отображать разные размеры и/или цвета для каждой точки, поэтому рендерер должен выполнять дополнительную работу по построению каждой точки отдельно.
В ``plt.plot``, с другой стороны, точки всегда по существу являются копиями друг друга, поэтому работа по определению внешнего вида точек графика выполняется только один раз для всего набора данных.
Для больших наборов данных разница между этими двумя методами может привести к значительному различию в производительности, и по этой причине для больших наборов данных ``plt.plot`` следует предпочесть ``plt.scatter``.
